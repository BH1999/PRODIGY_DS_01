Task - 1:
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv'
population_data = pd.read_csv(file_path, skiprows=4)  # Skip initial non-data rows

# Selecting data for a specific year, let's take the most recent year available, 2023
year = '2023'
population_2023 = population_data[['Country Name', year]].dropna()

# Plotting a bar chart of population distribution in 2023 for the top 10 most populous countries
top_10_countries = population_2023.nlargest(10, year)

plt.figure(figsize=(12, 8))
plt.bar(top_10_countries['Country Name'], top_10_countries[year], color='skyblue')
plt.title('Top 10 Most Populous Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
